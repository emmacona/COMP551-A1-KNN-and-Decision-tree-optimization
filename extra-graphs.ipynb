```python
import matplotlib.pyplot as plt
import numpy as np

x = [10, 20, 30, 100]
knn_d1 = [0.818,
0.827,
0.831,
0.845,]
dt_d1 = [0.825,
0.846,
0.847,
0.854,
]

knn_d2 = [0.996,
0.998,
0.998,
0.998,
]
dt_d2 = [0.997,
0.998,
0.998,
0.998,
]

fig = plt.figure(figsize=(15, 5))
ax = fig.add_subplot(111)    # The big subplot
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.plot(x, knn_d1, label='KNN') 
ax1.plot(x, dt_d1, label='Decision Tree') 
ax1.set_title("Accuracy of models on growing subsets of dataset 1")
ax1.legend()
ax1.grid()
ax1.set_axis_on()

ax2.plot(x, knn_d2, label='KNN') 
ax2.plot(x, dt_d2, label='Decision Tree') 
ax2.set_title("Accuracy of models on growing subsets of dataset 2")
ax2.legend()
ax2.grid()
ax2.set_axis_on()
ax1.set_ylabel('Accuracy on applied model')
ax1.set_xlabel('Percentage of data used')
ax2.set_ylabel('Accuracy on applied model')
ax2.set_xlabel('Percentage of data used')
ax.set_axis_off()

```
